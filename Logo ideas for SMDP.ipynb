```python
%pylab inline
import numpy as np
import matplotlib.font_manager
import matplotlib.pyplot as plt
matplotlib.font_manager.findSystemFonts()
```

```python
numCircles = 7
overlap = 75
rad = 0.3
alp = 0.05
font = {'family': 'sans',
        'color':  '#DDFFDD',
        'weight': 'normal',
        'style': 'normal',
        'size': 300,
        }
fig, ax = plt.subplots(figsize=(8,8)) # note we must use plt.subplots, not plt.subplot
for a in range(numCircles):
    circle1 = plt.Circle((0.5+10.0/overlap*np.cos(a*2.0*np.pi/numCircles), 0.5+10.0/overlap*np.sin(a*2.0*np.pi/numCircles)), rad,
                         color='k', alpha=alp+0.15*(1-np.sin(a*2.0*np.pi/numCircles)))
    ax.add_artist(circle1)
ax.text(0.5,0.47,'A',horizontalalignment='center',verticalalignment='center',alpha=0.6,fontdict=font)
ax.set_aspect('equal')
```
